2. Pie chart perbandingan banyaknya masing-masing neighbourhood dengan harga rumah per malam di
atas 300.

```python
import pandas as pd
import matplotlib.pyplot as plt

df=pd.read_csv("P05-03.csv")

kost=df.loc[df["price"]>300]
print(kost)

kost["neighbourhood"].value_counts().plot(kind="pie")
plt.tight_layout()
```

3. Scatter plot dengan long sebagai x dan lat sebagai y dari rumah yang ada di neighbourhood Kallang.

```python
import pandas as pd
import matplotlib.pyplot as plt

df=pd.read_csv("P05-03.csv")
kallang=df.loc[df["neighbourhood"]=="Kallang"]
print(kallang)

kallang.plot(kind="scatter", x="longitude", y="latitude")
```

4. Bubble plot dengan long sebagai x, lat sebagai y, dan banyaknya review sebagai ukuran dari rumah
dengan banyak hari ketersediaan rumah dalam satu tahun di atas 100

```python
import pandas as pd
import matplotlib.pyplot as plt

df=pd.read_csv("P05-03.csv")
ready=df.loc[df["availability_365"]>100]

ready.plot(kind="scatter", x="longitude", y="latitude", sizes=ready["number_of_reviews"])
```
