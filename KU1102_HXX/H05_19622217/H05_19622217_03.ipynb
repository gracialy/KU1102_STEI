```python
import pandas as pd
import matplotlib.pyplot as plt

# membaca file
df=pd.read_csv("H05.csv")
```

1. histogram distribusi nilai matematika

```python
df["nilai_mat"].plot(kind="hist", bins=[i for i in range (0,100,1)], color="#f78352")
```

2. Diagram batang horizontal banyaknya mahasiswa masing-masing fakultas

```python
df["fakultas"].value_counts().plot(kind="barh", title="banyak mahasiswa", color="#f78352")

# menampilkan banyak mahasiswa di samping setiap bar
for index, value in enumerate(df["fakultas"].value_counts()):
    plt.text(value, index, str(value))
```

3. Diagram pie banyaknya mahasiswa masing-masing fakultas

```python
# autopct: menampilkan persen mahasiswa sampai 2 desimal
df["fakultas"].value_counts().plot(kind="pie", autopct="%.2f%%")
# untuk memperbaiki layout axis dan label
plt.tight_layout()
```

4. Berdasarkan diagram pie dan batang horizontal, manakah fakultas dengan mahasiswa terbanyak? 
Manakah diagram yang lebih baik dalam menampilkan fakultas dengan mahasiswa terbanyak dan mengapa?

Menurut saya, menampilkan fakultas dengan jumlah mahasiswa terbanyak paling baik digunakan dengan diagram batang horizontal:
diagram batang horizontal lebih mudah dilihat jurusan mana yang paling banyak berdasarkan panjang batang,
sedangkan diagram pie perlu diamati dengan seksama besar pembagian untuk melihat bagian terbesar.
Tampilan banyak siswa juga lebih cocok pada bar horizontal, sedangkan pie chart digunakan untuk memvisualisasi persenan

5. Scatter plot dengan nilai kimia sebagai x dan fisika sebagai y

```python
df.plot(kind="scatter", x="nilai_kim", y="nilai_fis", color="#f78352")
```
