1. Ada berapa rumah di Porto Alegre dengan luas di atas 150?

```python
import pandas as pd

# KAMUS
# df: dataframe

df=pd.read_csv("P05-02.csv")

# test read
# pd.read_csv("P05-02.csv")

# mengambil data rumah di Porto Alegre dengan luas di atas 150
# df.loc[(df["city"]=="Porto Alegre") & (df["area"]>150)]
# kemudian menampilkan banyak rumah yang memenuhi syarat
len(df.loc[(df["city"]=="Porto Alegre") & (df["area"]>150)])
```

2. Data rumah dengan pajak bangunan per tahun terbesar.

4. Banyak rumah yang sudah dilengkapi furnitur (furnished) untuk masing-masing kota.

```python
import pandas as pd

# KAMUS
# df: dataframe

df=pd.read_csv("P05-02.csv")

# mengambil data rumah yang sudah dilengkapi furnitur
# df.loc[df["furniture"]=="furnished"]
# mendaftar kemunculan tiap kota pada data yang memenuhi syarat
df.loc[df["furniture"]=="furnished"]["city"].value_counts()
```

5. 5 data rumah dengan harga sewa total per tahun termurah yang berada di kota Campinas.

```python
import pandas as pd

# KAMUS
# df: dataframe

df=pd.read_csv("P05-02.csv")

# mengambil data rumah yang berada di kota Campinas
# df.loc[df["city"]=="Campinas"]
# mengurutkan data yang memenuhi syarat berdasarkan harga sewa total per tahun, menaik
df.loc[df["city"]=="Campinas"].sort_values(["total"],ascending=[True])[:5]
```
